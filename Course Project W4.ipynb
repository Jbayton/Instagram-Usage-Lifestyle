```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('instagram_usage_lifestyle.csv')

plt.figure(figsize=(7, 7))
plt.scatter(df['daily_active_minutes_instagram'], df['body_mass_index'])

plt.xlabel('Daily Minutes on Instagram')
plt.ylabel('Body Mass Index (BMI')
plt.title('Relationship Between BMI and Screen Time on IG')

plt.show()
```

```python
df['BMI_Category'] = pd.cut(
    df['body_mass_index'],
    bins=[0, 18.5, 25, 30, df['body_mass_index'].max()],
    labels=['Underweight', 'Normal', 'Overweight', 'Obese'])

df['Minutes_Category'] = pd.cut(
    df['daily_active_minutes_instagram'],
    bins=[0, 10, 30, df['daily_active_minutes_instagram'].max()],
    labels=['Low', 'Medium', 'High'])

contingency_table = pd.crosstab(
    df['BMI_Category'],
    df['Minutes_Category'])

print(contingency_table)

from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-square statistic:", chi2)
print("p-value:", p)
print("Degrees of freedom:", dof)
```
